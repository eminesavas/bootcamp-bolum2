```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import warnings
warnings.filterwarnings('ignore')
```

## Soru 1

```python
ogrenci_performans = pd.read_csv('data/StudentsPerformance (1).csv')
ogrenci_performans
```

```python
plt.figure(1)
plt.subplot(221)
ogrenci_performans['gender'].value_counts(normalize=True).plot.bar(figsize=(14,14), title='Cinsiyet')

plt.subplot(222)
ogrenci_performans['test preparation course'].value_counts(normalize=True).plot.bar(title = 'Hazırlık Kursu')

plt.subplot(223)
ogrenci_performans['race/ethnicity'].value_counts(normalize=True).plot.bar(title='Sınıf')

plt.subplot(224)
ogrenci_performans['parental level of education'].value_counts(normalize=True).plot.bar(title = 'Ebeveyn Eğitim Seviyesi')


plt.show()
```
